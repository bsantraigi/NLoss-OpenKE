```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def logsigmoid(x):
    return -np.log(1+np.exp(-x))
```

```python
gamma = 5
```

```python
def DistMult(head, relation, tail):
#         score = gamma - np.linalg.norm(head+relation-tail, axis = 0)
        score = np.sum(head*relation*tail, axis = 0)
        return score
```

```python
# for _ in range(20):
d = 10
def S():
    # return np.random.randn(d)
    return np.random.gamma(2,0.7,d)
h1 = S()
t1 = S()
r1 = np.random.randn(d)

h2 = S()
t2 = S()
r2 = np.random.randn(d)

x = []
y = []
ly  = []
for t in np.arange(0,1,0.01):
    try:
        u = DistMult(h1+t*h2,r1+t*r2,t1+t*t2)
        v = logsigmoid(u)
    except OverflowError as e:
        continue
    x.append(t)
    y.append(u)
    ly.append(v)

fig, ax = plt.subplots(1,2, figsize=(10,4))
ax[0].plot(x, y)
ax[1].plot(x, ly)
ax[0].legend(['R'])
ax[1].legend(['logsigmoid(R)'])
```
