```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def logsigmoid(x):
    return -np.log(1+np.exp(-x))
```

```python
gamma = 5
```

```python
def TransE(head, relation, tail):
        score = gamma - np.linalg.norm(head+relation-tail, axis = 0)
        return score
```

```python
d = 10
h1 = np.random.randn(d)
t1 = np.random.randn(d)
r1 = np.random.randn(d)

h2 = np.random.randn(d)
t2 = np.random.randn(d)
r2 = np.random.randn(d)

x = []
y = []
ly  = []
for t in np.arange(-10,10,0.01):
    try:
        u = TransE(h1+t*h2,r1+t*r2,t1+t*t2)
        v = logsigmoid(u)
    except OverflowError as e:
        continue
    x.append(t)
    y.append(u)
    ly.append(v)

fig, ax = plt.subplots(1,2, figsize=(10,4))
ax[0].plot(x, y)
ax[1].plot(x, ly)
ax[0].legend(['R'])
ax[1].legend(['logsigmoid(R)'])
```
